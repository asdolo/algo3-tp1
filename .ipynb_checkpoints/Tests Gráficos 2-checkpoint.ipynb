```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
from matplotlib.ticker import MaxNLocator
```

<h3>Complejidad Fuerza Bruta</h3>

```python
df_complejidad_fb = pd.read_csv('output/mediciones-fuerzabruta.csv')
del df_complejidad_fb['resultado']

# Obtengo el promedio
df_complejidad_fb = df_complejidad_fb.groupby(['n', 'V'], as_index=False, sort=False).mean()

# grafico
a4_dims = (10.75, 3.975)
fig, ax = plt.subplots(figsize=a4_dims)
sns.set_palette(sns.color_palette("cubehelix", 1))
sns.set_style("whitegrid")
ax.xaxis.set_major_locator(MaxNLocator(integer=True))

df_complejidad_fb['fb/n*2^n'] = df_complejidad_fb['tiempo'] / (df_complejidad_fb['n'] * (2**df_complejidad_fb['n']));

ax1 = sns.lineplot(x="n", y="fb/n*2^n", data=df_complejidad_fb, ax=ax);
plt.ylim(-0.05,0.05);

plt.ylabel(r"Relaci칩n $\frac{FuerzaBruta(n, V)}{n\times 2^{n}}$", fontsize=16);
plt.xlabel(r"$n$", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
```

<h3>Complejidad Backtracking</h3>

```python
df_complejidad_bt = pd.read_csv('output/mediciones-backtracking.csv')
del df_complejidad_bt['resultado']

# Obtengo el promedio
df_complejidad_bt = df_complejidad_bt.groupby(['n', 'V'], as_index=False, sort=False).mean()

# grafico
a4_dims = (10.75, 3.975)
fig, ax = plt.subplots(figsize=a4_dims)
sns.set_palette(sns.color_palette("cubehelix", 1))
sns.set_style("whitegrid")
ax.xaxis.set_major_locator(MaxNLocator(integer=True))

df_complejidad_bt['bt/n*2^n'] = df_complejidad_bt['tiempo'] / (df_complejidad_bt['n'] * (2**df_complejidad_bt['n']));

ax1 = sns.lineplot(x="n", y="bt/n*2^n", data=df_complejidad_bt, ax=ax);
plt.ylim(-0.05,0.05);

plt.ylabel(r"Relaci칩n $\frac{Backtracking(n, V)}{n\times 2^{n}}$", fontsize=16);
plt.xlabel(r"$n$", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
```

<h3>Complejidad Programaci칩n Din치mica</h3>

```python
df_complejidad_pd = pd.read_csv('output/mediciones-programaciondinamica.csv')
del df_complejidad_pd['resultado']

# Obtengo el promedio
df_complejidad_pd = df_complejidad_pd.groupby(['n', 'V'], as_index=False, sort=False).mean()

# grafico
a4_dims = (10.75, 3.975)
fig, ax = plt.subplots(figsize=a4_dims)
sns.set_palette(sns.color_palette("cubehelix", 1))
sns.set_style("whitegrid")
ax.xaxis.set_major_locator(MaxNLocator(integer=True))

df_complejidad_pd['n*V'] = df_complejidad_pd['n'] * df_complejidad_pd['V']
df_complejidad_pd['pd/n*V'] = df_complejidad_pd['tiempo'] / (df_complejidad_pd['n'] * df_complejidad_pd['V'])

ax1 = sns.lineplot(x="n*V", y="pd/n*V", data=df_complejidad_pd, ax=ax)
plt.ylim(-0.1,0.10);

plt.ylabel(r"Relaci칩n $\frac{Programaci칩nDin치mica(n, V)}{n\times V}$", fontsize=16)
plt.xlabel(r"$n \times V$", fontsize=16)
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
```

<h1>Experimento 1</h1>
Fuerza Bruta vs Backtracking (sin podas)

```python
df_test_1_fb = pd.read_csv('output/test-1-fuerzabruta.csv')
del df_test_1_fb['resultado']

df_test_1_bt = pd.read_csv('output/test-1-backtracking.csv')
del df_test_1_bt['resultado']

# Obtengo el promedio en cada tabla
df_test_1_fb = df_test_1_fb.groupby(['n', 'V'], as_index=False, sort=False).mean()
df_test_1_bt = df_test_1_bt.groupby(['n', 'V'], as_index=False, sort=False).mean()

# renombro las columnas en cada dataframe y los mergeo en uno solo
df_test_1_fb.columns = ['n', 'V', 'tiempo (fuerzabruta)']
df_test_1_bt.columns = ['n', 'V', 'tiempo (backtracking)']

df_test_1 = df_test_1_fb
df_test_1['tiempo (backtracking)'] = df_test_1_bt['tiempo (backtracking)']

# grafico
a4_dims = (6.75, 3.975)
fig, ax = plt.subplots(figsize=a4_dims)
sns.set_palette(sns.color_palette("cubehelix", 2))
sns.set_style("whitegrid")
ax.xaxis.set_major_locator(MaxNLocator(integer=True))


ax1 = sns.lineplot(x="n", y="tiempo (fuerzabruta)", data=df_test_1, label='Fuerza Bruta', ax=ax);
ax1 = sns.lineplot(x="n", y="tiempo (backtracking)", data=df_test_1, label='Backtracking', ax=ax);


plt.ylabel("Tiempo de ejecuci칩n [ms]", fontsize=16);
plt.xlabel("n", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
```

<h3>Poda de medio 치rbol</h3>
Complejidad practica esperada: O(n*2^n/2)

```python
df_test_2_a_fb = pd.read_csv('output/test-2-a-fuerzabruta.csv')
del df_test_2_a_fb['resultado']

df_test_2_a_bt = pd.read_csv('output/test-2-a-backtracking.csv')
del df_test_2_a_bt['resultado']

# Obtengo el promedio en cada tabla
df_test_2_a_fb = df_test_2_a_fb.groupby(['n', 'V'], as_index=False, sort=False).mean()
df_test_2_a_bt = df_test_2_a_bt.groupby(['n', 'V'], as_index=False, sort=False).mean()

# renombro las columnas en cada dataframe y los mergeo en uno solo
df_test_2_a_fb.columns = ['n', 'V', 'tiempo (fuerzabruta)']
df_test_2_a_bt.columns = ['n', 'V', 'tiempo (backtracking)']

df_test_2_a = df_test_2_a_fb
df_test_2_a['tiempo (backtracking)'] = df_test_2_a_bt['tiempo (backtracking)']

# grafico
a4_dims = (6.75, 3.975)
fig, ax = plt.subplots(figsize=a4_dims)
sns.set_palette(sns.color_palette("cubehelix", 3))
sns.set_style("whitegrid")
ax.xaxis.set_major_locator(MaxNLocator(integer=True))

ax1 = sns.lineplot(x="n", y="tiempo (fuerzabruta)", data=df_test_2_a, label='Fuerza Bruta', ax=ax);
ax1 = sns.lineplot(x="n", y="tiempo (backtracking)", data=df_test_1, label='Backtracking (Sin podas - Experimento 1)', ax=ax);
ax1 = sns.lineplot(x="n", y="tiempo (backtracking)", data=df_test_2_a, label='Backtracking (Poda de medio 치rbol)', ax=ax);


plt.ylabel("Tiempo de ejecuci칩n [ms]", fontsize=16);
plt.xlabel("n", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
```
